```python

scenarios_list = ["best", "# Orders",  "Design Completion", "A/E Proficiency", "Construction Proficiency",\
        "Supply Chain Proficiency", "Design Maturity", "Standardization", "Modularization",\
        "Commercial BOP", "Safety-Related RB", "ITC", "Interest Rate"]

for reactor_type in reactor_type_list:
    print(reactor_type)
    
    TCI_list_avg = []
    
    for scenario in scenarios_list:
        print(scenario)
        
        if scenario == '# Orders':
            num_orders = 2
        else:   
            num_orders = 20 
        
        if scenario == "Startup Duration":
                startup_0 = 36 # start up duration (months)
        else:
                startup_0 = 12
        # land cost
            # From the SA report: the cost $22,000 per acre. The land area is 500 acres including recommended buffer
        if scenario == "Land Cost":
                land_cost_per_acre_0 =  100000 # d
        else:  
                land_cost_per_acre_0 =  1000 # dollars/acre   
  

        # # interest rate :
        if scenario == "Interest Rate":
            interest_rate_0 = 0.12
        else:    
            interest_rate_0 = 0.04

        if scenario == "Design Completion":
            design_completion_0 = 0.5 # 1 means 100%
        else: 
            design_completion_0 = 1# 1 means 100%
        
        if scenario == "Design Maturity":
            Design_Maturity_0 = 0
        else:   
                Design_Maturity_0 = 2
                 
                 
                 
                 
                 # #procurement service experience (supply chain experience)
        
        
        if scenario == "Supply Chain Proficiency":
            proc_exp_0= 0 # 2 means procurement experts. This is ideal. 
        else:
            proc_exp_0= 2
                
        # #  architecture and engineeringexperience
        if scenario == "A/E Proficiency":
            ae_exp_0 = 0
        else:    
            ae_exp_0 = 2
                
                
        if scenario == "Construction Proficiency":
        # #  Construction service experience
            ce_exp_0 = 0
        else:   
            ce_exp_0 = 2

        if scenario == "Modularization":
        # modularity (applied on civil construction only) "stick_built"  or "modularized"
            mod_0 = "stick_built" 
        else:  
            mod_0 = "modularized"

        
        if scenario == "Standardization":
        # cross_site_standardization :
            standardization_0 = 0.7 # 0.7 corresponds to 70% standardization for PWRs
        else:
            standardization_0 = 0.95
        
            
        if scenario == "Commercial BOP":
        # # Determining if the BOP and reactor building (containtment) are non-nuclear or nuclear grade equipment (safety related)
            BOP_grade_0 = "nuclear"
        else:    
            BOP_grade_0 = "non_nuclear"
        
        
        if scenario == "Safety-Related RB":
            RB_grade_0 = "nuclear"
        else:    
                RB_grade_0 = "non_nuclear"


        if scenario == "ITC":
        # #investment tax credits subsidies
            ITC_0 =  0 #
            
            #number of reactors claiming ITC
            n_ITC = 3 
        else:  
            ITC_0 =  0.4 #
            n_ITC = 3
        
        TCI_list = [] 
        for n_th in range(1, num_orders+1):
            results = calculate(num_orders ,n_th, startup_0, interest_rate_0, design_completion_0, Design_Maturity_0, proc_exp_0 , ae_exp_0, ce_exp_0, mod_0 , standardization_0, BOP_grade_0, RB_grade_0, ITC_0, n_ITC, reactor_type )
            
            TCI_result = results[1]
            TCI_list.append( TCI_result )
        avg_TCI = np.mean(TCI_list)
        
        TCI_list_avg.append(avg_TCI)

    if reactor_type == "SFR":
        TCI_list_avg_SFR =  TCI_list_avg
    
    elif reactor_type == "HTGR":
        TCI_list_avg_HTGR      =  TCI_list_avg 
```
